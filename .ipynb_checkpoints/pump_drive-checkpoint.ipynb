## **Splitting Dataset into training and testing by train_test_split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=22)
```

## ***1.Logistic Regression Model***

```python
from sklearn.linear_model import LogisticRegression
model_lgr=LogisticRegression()
model_lgr.fit(X_train,y_train)
```

```python
y_predict_lgr=model_lgr.predict(X_test)
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy of the model by Logistic Regression is :",accuracy_score(y_test,y_predict_lgr)*100)
print(confusion_matrix(y_test,y_predict_lgr))
```

```python
print(classification_report(y_test,y_predict_lgr))
```

```python
import pickle

with open("model_lgr.pkl", "wb") as f:
    pickle.dump(model_lgr, f)
```

## ***2.KNN***

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier()
```

```python
model_knn.fit(X_train,y_train)
```

```python
y_predict_knn=model_knn.predict(X_test)
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy of the model by KNN is :",accuracy_score(y_test,y_predict_knn)*100)
```

```python
print(confusion_matrix(y_test,y_predict_knn))
```

```python
print(classification_report(y_test,y_predict_knn))
```

```python
with open("model_knn.pkl", "wb") as f:
    pickle.dump(model_lgr, f)
```

## ***3.Decision Tree*** ##

```python
from sklearn import tree

model_tree = tree.DecisionTreeClassifier()   # empty model of the decision tree
model_tree.fit(X_train,y_train)
```

```python
y_predict_tree=model_tree.predict(X_test)
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy of the model by Decision Tree is : ",accuracy_score(y_test,y_predict_tree)*100)
```

```python
print(classification_report(y_test,y_predict_tree))
```

```python
import csv
from datetime import datetime
import os

def log_data(soil, humidity, temperature, aqi, pump_status, filename="logs.csv"):
    """
    Appends sensor readings with timestamp to a CSV log file.

    Parameters:
        soil (float): Soil moisture level
        humidity (float): Humidity percentage
        temperature (float): Temperature in C
        aqi (int): Air Quality Index
        pump_status (str): 'ON' or 'OFF'
        filename (str): Path to the CSV file (default 'logs.csv')
    """
    # Get current timestamp
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Check if file exists, to decide if header is needed
    file_exists = os.path.isfile(filename)

    # Open file in append mode
    with open(filename, mode='a', newline='') as file:
        writer = csv.writer(file)

        # Write header only if file is new
        if not file_exists:
            writer.writerow(["Timestamp", "Soil", "Humidity", "Temperature", "AQI", "Pump_Status"])

        # Append data row
        writer.writerow([timestamp, soil, humidity, temperature, aqi, pump_status])

    print(f" Log added: {timestamp}, Soil={soil}, Humidity={humidity}, Temp={temperature}, AQI={aqi}, Pump={pump_status}")
```

```python
port = "COM9"
baud = 115200

"""
<data>{'soil":"0"'aqi":"245"'temp":"25.80"'hum":"37"</data>}
"""

serialPort = serial.Serial(port, baud, timeout=1)

time.sleep(2)

if serialPort.isOpen():
    print(serialPort.name + ' is open...')
progress = 1

if selected_algo == "":
    raise Exception("Plsease select a aloritham!")

print(selected_algo)
    
try:
    label_pump = widgets.Label(value="0")
    label_soil =  widgets.Label(value="0")
    label_temp =  widgets.Label(value="0")
    label_aqi =  widgets.Label(value="0")
    label_hum =  widgets.Label(value="0")
    
    display(label_pump)
    display(label_soil)
    display(label_temp)
    display(label_aqi)
    display(label_hum)
    
    #label_pump.value = f"OFF: {0}"

    while True:
        in_data = serialPort.readline()
        # print('in_data', in_data)
        try:
            in_str = in_data.decode('utf-8')
            #print('in_str', in_str)

            if "<data>" in in_str and "</data>" in in_str:
                refined_json_str  =  in_str.split("<data>")[1]
                refined_json_str = refined_json_str.split("</data>")[0]
                #print('refined_json_str', refined_json_str)

                refined_json_str = refined_json_str.strip()
                json_data = json.loads(refined_json_str)
                #print(json_data)
                soil = float(json_data['soil'])
                aqi = float(json_data['aqi'])
                temp = float(json_data['temp'])
                hum = float(json_data['hum'])
                
                
                
                
                
                #['temperature', 'humidity', 'soil_moisture',
                
                    
                state = getPumpStatePredection(temp, hum, soil)
#                 print(state)
                pump_cmd = 'PUMP_ON'
                
                if state == 0:
                    pump_cmd = 'PUMP_OFF'
                

                prep = ""
                if progress == 1:
                    prep = "|"
                elif progress == 2:
                    prep = "/"
                elif progress == 3 or progress == 4 :
                    prep = "-"
                elif progress == 5:
                    prep = "\\"
                    
                if soil == 1.0:
                    soil = "WET"
                else:
                    soil = "DRY"
                
                label_pump.value = "Pump:"+pump_cmd + " "+prep
                label_soil.value =  "Soil:"+soil
                label_temp.value =  "Temp:"+ str(temp)
                label_aqi.value =   "Aqi:"+str(aqi)
                label_hum.value =   "Hum:"+str(hum)
                
                log_data(soil, hum, temp, aqi, pump_cmd)
                progress+=1
                if progress >= 6:
                    progress = 1
                    
               
                try:
                    cmd_tosend= pump_cmd+"\n"
                    serialPort.write(cmd_tosend.encode('utf-8'))
                except Exception as e:
                    print(e)
        except Exception as e:
            print(e)
except KeyboardInterrupt:
    print("\nKeyboardInterrupt caught! closing port")
    serialPort.close()
```
